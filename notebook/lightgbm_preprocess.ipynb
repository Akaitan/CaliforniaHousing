### サンフランシスコとロサンゼルスに近い緯度経度の部分にフラグを立てる。

```python
sanf_plot = [37.76667, -122.4333]
los_plot = [34.0522, -118.2436]
```

```python
# フラグを作ったがあまり効果はない。
for i in range(len(data_exp)):
    if abs(data_exp["Latitude"].iloc[i] - sanf_plot[0]) <= 0.1:
        data_exp.at[i, "san_lattitude_flg"] = 1
    else:
        data_exp.at[i,"san_lattitude_flg"] = 0
    
    if abs(data_exp["Latitude"].iloc[i] - los_plot[0])<= 0.1:
        data_exp.at[i, "los_lattitude_flg"] = 1
    else:
        data_exp.at[i, "los_lattitude_flg"] = 0
    
    # 経度
    if abs(data_exp["Longitude"].iloc[i] - sanf_plot[1])<= 0.1:
        data_exp.at[i, "san_longitude_flg"] = 1
    else:
        data_exp.at[i, "san_longitude_flg"] = 0

    if abs(data_exp["Longitude"].iloc[i] - los_plot[1])<= 0.1:
        data_exp.at[i, "los_longitude_flg"] = 1
    else:
        data_exp.at[i, "los_longitude_flg"] = 0
data_exp
```

```python
import math
for i in range(len(data_exp)):

    # サンフランシスコからの距離
    lati_san = (data_exp["Latitude"].iloc[i] - sanf_plot[0]) ** 2 
    longi_san = (data_exp["Longitude"].iloc[i] - sanf_plot[1]) ** 2
    data_exp.at[i, "distance_of_san"] = math.sqrt(lati_san + longi_san)

    # ロサンゼルスからの距離
    lati_los = (data_exp["Latitude"].iloc[i] - los_plot[0]) ** 2 
    longi_los = (data_exp["Longitude"].iloc[i] - los_plot[1]) ** 2
    data_exp.at[i, "distance_of_los"] = math.sqrt(lati_los + longi_los)
```

```python
data_exp
```

```python
%matplotlib inline
import graphviz
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
# from sklearn.tree import DecisionTreeRegressor
import lightgbm as lgb
```

```python
x_train, x_test, y_train, y_test = train_test_split(data_exp,
                                                    data_target,
                                                    test_size = 0.2,
                                                    random_state = 0)
```

```python
lgb_train = lgb.Dataset(x_train, y_train)
lgb_valid = lgb.Dataset(x_test, y_test)  
```

```python
params = {'boosting_type':'gbdt',
          'objective': 'regression',
          'max_bin':300,
          'learning_rate':0.05,
          'num_leaves':31,
          'metric' :'mse'
         }
verbose_eval = 0
```

```python
model = lgb.train(params = params,
                  train_set = lgb_train,
                  valid_sets = lgb_valid,
                  num_boost_round = 100,
                    callbacks=[lgb.early_stopping(stopping_rounds=10, 
                    verbose=True), # early_stopping用コールバック関数
                    lgb.log_evaluation(verbose_eval)] # コマンドライン出力用コールバック関数
                )
```

```python
# 訓練データにpredictさせてみる。
y_train_pred = model.predict(x_train,
                             num_iteration = model.best_iteration)
```

```python
# テストデータの予測を行う。
y_test_pred = model.predict(x_test,
                            num_iteration = model.best_iteration)
```

```python
y_test_pred
```

```python
# trainデータのMSE
y_train_arrays = y_train.to_numpy()
mean_squared_error(y_train_arrays.flatten(), y_train_pred, squared = True)
```

```python
# testデータのMSE
y_test_arrays =y_test.to_numpy()
mean_squared_error(y_test_arrays.flatten(), y_test_pred, squared = True)
```

```python
# 念のためnumpyでもtestデータのmseを計算しておく。
# 0.653311…で一致。
y_test_array =y_test.to_numpy()

# レイべリングして一次元に変換。
y_test_array = y_test_array.flatten()
mse_numpy = np.mean((y_test_array- y_test_pred) ** 2)
print(mse_numpy)
```

### 結果
 - 前処理 -> 決定木でのモデル作成 -> 評価
 - MSEは0.22853（フラグなし）
 - MSEは（フラグあり）
 - 圧倒的に精度が高い。
 - カリフォルニアとサンフランシスコの近くのフラグを作った。若干MSEは減少したが、まだやりようはありそう。
